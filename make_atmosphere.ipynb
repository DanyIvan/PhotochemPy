```python
fil = open('input/in_2.dist','r')
lines = fil.readlines()
nq = 61
k_bolt = 1.380649e-16
atmos = np.zeros([nq,200])


# now lets 
dz = 0.5e5
Alt =np.arange(dz-.5*dz,100e5,dz)


a = 0
j = 0


for i in range(nq):
    a =  int((i)/10)
    if not i % 10:
        j = 0
    for k,line in enumerate(lines[200*a:200*a+200]):
        atmos[i,k] = float(line.split()[j])
    j+=1
    

T=np.zeros(200)
EDD=np.zeros(200)
DEN=np.zeros(200)
O3=np.zeros(200)
nCO2=np.zeros(200)
for i,line in enumerate(lines[1400:1600]):
    T[i],EDD[i],DEN[i],O3[i],nCO2[i] = [float(aa) for aa in line.split()]
    

aersol = []
wfall = []
rpar = []
for i,line in enumerate(lines[1600:1800]):
    aersol.append([float(aa) for aa in line.split()][0:4])
    wfall.append([float(aa) for aa in line.split()][4:8])
    rpar.append([float(aa) for aa in line.split()][8:12])

aersol = np.array(aersol)
wfall = np.array(wfall)
rpar = np.array(rpar)
    
    
    
fil.close

fil = open('input/speciesOG.dat','r')

lines = fil.readlines()

species = []


for line in lines[18:18+nq]:
    species.append(line.split()[0])
    
    
# make dict
f = {}
f['alt'] = Alt
f['press'] = k_bolt*T*DEN/1e6
f['temp'] = T
f['density'] = DEN
f['eddy'] = EDD
for i,spec in enumerate(species):
    if spec=='SO2':
        print(i)
    f[spec] = atmos[i]
    
f['CO2'] = nCO2/DEN

f['N2'] = 1-(np.sum(atmos,axis=0)+nCO2/DEN)


particles  = ['SO4AER','S8AER','HCAER','HCAER2']
params = ['AERSOL','WFALL','RPAR']

for j in range(len(params)):
    for i in range(len(particles)):
        if params[j] == 'AERSOL':
            f[particles[i]+'_'+params[j]] =  aersol[:,i]
        if params[j] == 'WFALL':
            f[particles[i]+'_'+params[j]] =  wfall[:,i]
        if params[j] == 'RPAR':
            f[particles[i]+'_'+params[j]] =  rpar[:,i]
```

```python
def profile2dic(filename):
    file = open(filename)
    lines = file.readlines()
    key = lines[0].split()
    #build dictionary of output
    out = []
    for i in range(0,len(lines)):
        if i>0:
            tmp = []
            for j in lines[i].split():
                try:
                    tmp.append(float(j))
                except ValueError:
                    tmp.append(0)
            out.append(tmp)
    out = np.array(out)

    f_out = {}
    for i in range(0,len(key)):
        f_out[key[i]]=out[:,i]
    file.close()

    return f_out
```
